## 环境安装教程



下载anaconda 安装(just me选项） 
安装后添加环境变量 python （jupyter notebook路径是否添加可选）
D:\Anaconda\Scripts\   
D:\Anaconda\
升级pip3 :python.exe -m pip install --upgrade pip
添加进环境变量pip3 


查看版本号:Pip3 list pip freeze
查看需要更新的库:pip list --outdated 可以简写 pip list -o
更新库：pip install --upgrade 库名
卸载库：pip uninstall 库名
更改jupyter的默认目录 找到启动文件 右键属性 修改起始位置为你想要的文件夹 然后将目标中的`%USERPROFILE%`去掉就好了。

更改jupyter的主题：安装命令pip install jupyter-themes  可选主题：jt -l   选择主题：jt -t oceans16 -T -N  
代码折叠等功能：https://blog.csdn.net/August1226/article/details/80762632   
添加虚拟环境内核到jupyter中：在虚拟环境中 conda install nb_conda  

### TensorFlow环境安装
打开Anaconda Prompt，通过anaconda search -t conda tensorflow-gpu来查询有哪些tensorflow-gpu版本可以安装，从中找到符合自己设备配置的包进行安装
如：win10环境下安装1.13.1版本，要求python=3.7，会自动安装cuda10.0，cudnn7.3：
conda install tensorflow-gpu=1.13.1

### Pytorch环境安装
安装pytorch时如果网络状况良好，可以直接在pytorch官网上找到conda下安装对应环境的命令可以直接输入安装，如果在线安装网络不好，可以找到pip安装的命令下载对应的whl安装包，下载到本地再采用pip3进行安装。
如果是在虚拟环境中安装，需要将对应环境的pip.exe和pip-script.exe，改为condapip和condapip-script, 然后安装本地的包（为的的和环境变量下的pip进行区分），需要先安装torch，再安装torchvision。
另外需要安装cuda10.0和cudnn7.3，可以通过英伟达官网下载安装包进行安装，或通过anaconda  navigator 找到对应虚拟环境进行安装。
在python=3.7，cuda=10.0，cudnn=7.3环境下，tensorflow和pytorch可以共存。
### conda环境管理
|环境操作|命令|
|--------|----|
|查看环境|conda env list
|创建环境|conda create -n python36 python=3.6
|激活环境|source activate python36、activate python36 #windows下
|搜索包|conda search mxnet*
|指定环境，查看已安装的包|conda list -n python36
|指定环境，安装指定版本的包|conda install -n python36 mxnet==1.0.0
|指定环境，更新包|conda update -n python36 mxnet
|指定环境，删除包|conda remove -n python36 mxnet
|导出环境为yml|conda env export >  environment.yml
|根据yml创建环境|conda env create -f environment.yml
|对yml文件修改后更新环境|conda env update -f environment.yml
|退出环境|source deactivate、deactivate # windows下
|复制环境|conda create -n python36 --clone python36_new
|删除环境|conda remove -n python36 --all
更改镜像源
conda config --add channelsttps://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/
conda config --set show_channel_urls yes

## Jupyter Notebook 快捷键

从命令模式进入编辑模式需按 Enter 键，从编辑模式切换到命令模式需按 Esc 键。以下两表分别是对命令和编辑两种模式下快捷键的简单说明：

|Ctrl-Enter	| 运行本单元	||

|Y	|单元转入代码状态	
|M	|单元转入 markdown 状态	
|R	|单元转入 raw 状态	
|1	|设定 1 级标题	仅在 markdown 状态下时建议使用标题相关快捷键，如果单元处于其他状态，则会强制切换到 markdown 状态
|2	|设定 2 级标题	
|3	|设定 3 级标题	
|4	|设定 4 级标题	
|5	|设定 5 级标题	
|6	|设定 6 级标题
